```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
df = pd.read_csv('netflix_titles.csv')
print(df.head())
print(df.columns)
print(df.info())


#How many movies vs. TV shows are there?
movies_shows = df['type'].value_counts()
print("---movies vs tv shows---")
print(movies_shows)
#Which year had the most releases (release_year)?
most_releases = df['release_year'].value_counts().head(1)
print("---most releases---")
print(most_releases)
#How many shows have a missing director?
miss_dir = df['director'].isnull().sum()
print("---missing director---")
print(miss_dir)
#What are the top 5 countries with the most Netflix shows?
top_country = df['country'].value_counts().head(5)
print("---top 5 countries shows---")
print(top_country)
#What are the most common ratings?
most_rating = df['rating'].value_counts().idxmax()
print("---most comon ratings---")
print(most_rating)
#What are the top 5 directors (with the most titles listed)?
top_dir = print(df['director'].value_counts().head(5))
print("---top 5 directors with most titles---")
print(top_dir)
#What are the top 3 most common “listed_in” categories?
most_listed = df['listed_in'].value_counts().head(3)
print("---top 3 most common listed_in---")
print(most_listed)
#How many titles have “Documentary” in their “listed_in” field?
how_many = (df['listed_in'] == 'documentary').sum()
print("---titles and listed in---")
print(how_many)
#What’s the average release year of all titles?
averg_year = df['release_year'].mean()
print("---average release year for all titles---")
print(averg_year)
#Find the oldest title (earliest release_year) on Netflix:
oldest_title = df.loc[df['release_year'].idxmin()]
title = oldest_title['title']
year = oldest_title['release_year']
print("--- oldest title and year---")
print(title, year)

movies_shows = df['type'].value_counts()
plt.figure(figsize=(6, 4))
movies_shows.plot(kind='bar', color=['#4062bb', '#59c3c3'])
plt.title('Netflix: Movies vs TV Shows')
plt.ylabel('Count')
plt.xlabel('Type')
plt.tight_layout()
plt.savefig('movies_vs_tvshows.png')  # Save as image for Tableau or your portfolio!
plt.show()
```
